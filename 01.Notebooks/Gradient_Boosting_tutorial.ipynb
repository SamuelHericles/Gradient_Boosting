# Importando bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import ensemble
from sklearn import linear_model
```

## Configurando o tamanho da imagem

```python
plt.figure(figsize=(10, 9))
```

## Simulação de dados

```python
x = np.arange(0, 60)

def prod(x):
  y = x / 2 + (x // 10) % 2 * 20 * x / 5 + np.random.random() * 10
  return y

y = prod(x)
x = pd.DataFrame({'x': x})
```

```python
np.random.random() *10
```

## Plot dos dados

```python
plt.figure(figsize=(10, 5))
plt.scatter(x, y)
plt.show()
```

## Modelo de regressão linear

```python
linear_reg = linear_model.LinearRegression()
linear_reg.fit(x,y)

plt.figure(figsize=(10,9))


plt.subplot(2,1,1)
plt.title("Regressão Linear",fontsize=20)
plt.scatter(x,y)
plt.plot(x,linear_reg.predict(x),color='r')

plt.subplot(2,1,2)
plt.title("Resíduo do modelo de regressão linear",fontsize=20)
plt.scatter(x,y-linear_reg.predict(x),color='brown')

plt.show()
```

## Modelo Gradient Boosting

```python
#Definindo entrada e saída para o testo com o modelo
x = np.arange(0, 60)

def prod(x):
  y = x / 2 + (x // 10) % 2 * 20 * x / 5 + np.random.random() * 10
  return y

y = prod(x)
x = pd.DataFrame({'x': x})
```

```python
# Criação do modelo de gradient bossting, definindos os parâmetros
params = {
    'n_estimators': 10,
    'max_depth': 1,
    'learning_rate': 1,
    'criterion': 'mse'
}

gradient_boosting_regressor = ensemble.GradientBoostingRegressor(**params)
gradient_boosting_regressor.fit(x, y)


#Plot dos gráficos da saída produzida e do erro do modelo

plt.figure(figsize=(10, 9))

plt.subplot(2,1,1)
plt.title('Modelo Gradient Boosting',fontsize=20)
plt.scatter(x, y)
plt.plot(x, gradient_boosting_regressor.predict(x), color='r')

plt.subplot(2,1,2)
plt.title('Resíduos do modelo',fontsize=20)
plt.scatter(x, y - gradient_boosting_regressor.predict(x), color='brown')

plt.show()
```
